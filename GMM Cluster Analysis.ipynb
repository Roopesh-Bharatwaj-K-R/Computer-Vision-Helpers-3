```python
import pandas as pd
import numpy as np
# from luxvis import*
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score,accuracy_score
import glob
import os

def csv_path(path,path_a, ddir):
#     path= ('/home/roopesh/Desktop/CSV1/*.csv')
    for idx,csv_file in enumerate(glob.glob(path) ):
        input_csv_pixels= pd.read_csv(csv_file)

    # Inputing input_csv_pixels to GMM model
        from sklearn.mixture import GaussianMixture
        gm = GaussianMixture(n_components=3, random_state=0, covariance_type='full').fit(input_csv_pixels)

        cluster= gm.predict(input_csv_pixels)
        pred= pd.DataFrame(cluster)
        input_1=pd.DataFrame(input_csv_pixels)
        prediction=pd.concat([input_1,pred], axis=1)
        pred2 = pd.Series([idx for x in range(len(prediction.index))])
        prediction1=pd.concat([prediction, pred2],axis=1)
        c=('R_pix','G_pix','B_pix','GMM_cluster','Idx_csv')
        prediction1.to_csv(f'/home/roopesh/Documents/CSV2/{idx}.csv', sep=',',index=False,header=c)
        


if __name__ == '__main__':
    path= ('/home/roopesh/Desktop/CSV1/*.csv')
    csv_path(path)
```
